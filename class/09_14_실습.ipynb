```python
import nltk 
from nltk.corpus import gutenberg
story = gutenberg.open('melville-moby_dick.txt').read()

from nltk.tokenize import word_tokenize
tokens = word_tokenize(story)

# 
lemmatizer = nltk.stem.WordNetLemmatizer()  

process_tokens = tokens
#[lemmatizer.lemmatize(t) for t in process_tokens] #lemmatize 

stem = nltk.stem.PorterStemmer()
[stem.stem(t) for t in process_tokens] #stem

####################
word_count = dict()
for word in process_tokens:
    word_count[word] = word_count.get(word, 0) + 1
from wordcloud import WordCloud
wordcloud = WordCloud().generate_from_frequencies(word_count)

import matplotlib.pyplot as plt
plt.axis("off")
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```
